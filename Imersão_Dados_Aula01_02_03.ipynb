## Desafio 01 

```python

#Tabela_frequencia_ 1 (frequência tempo x dose)
#Metodo  value_counts 
tabela_frequencia_1 = dados.groupby(by=['tempo'])['dose'].value_counts() 
tabela_frequencia_1.unstack()
```

```python
#Tabela_frequencia_2 (frequência tempo e dose x tratamento)
tabela_frequencia_2 = dados.groupby(by=['tempo','dose'])['tratamento'].value_counts()
tabela_frequencia_2.unstack()
```

```python
# Tabela_frequencia_3 relativa usei parametro normalize=True (frequência tempo e dose x tratamento)
tabela_frequencia_2 = dados.groupby(by=['tempo','dose'])['tratamento'].value_counts(normalize=True)
tabela_frequencia_2.unstack()
```

```python
#Tabela_frequencia_4 (média do gene g-_0 agrupando por tempo, dose e tratamento)
tabela_frequencia_4 = dados.groupby(by=['tempo','dose','tratamento'])['g_0'].mean()
tabela_frequencia_4.unstack()
```

```python
#Tabela_frequencia_5 (max do gene g-_0 agrupando por tempo, dose e tratamento)
tabela_frequencia_5 = dados.groupby(by=['tempo','dose','tratamento'])['g_0'].max()
tabela_frequencia_5.unstack()
```
